```python
import pandas as pd 
data_participatiewet = pd.read_csv("Participatiewet_2018.csv", sep=';') 
data_participatiewet[['buurt_number', 'buurt_name']] = data_participatiewet['Buurten'].str.extract('(?P<buurt_number>[\d]+) (?P<buurt_name>[\s\S]+)', expand=True)
data_participatiewet = data_participatiewet.drop('Buurten', axis=1)
data_participatiewet = data_participatiewet.drop('buurt_name', axis=1)

print(data_participatiewet.shape)
print(data_participatiewet.columns)
data_participatiewet.head()
```

```python
data_dogs=pd.read_csv('honden_data.csv', sep = ";")
data_dogs = data_dogs.rename(columns={'buurtnr': 'buurt_number'})
data_dogs = data_dogs.rename(columns={'2018': 'dogs_amount'})
data_dogs['buurt_number'] = data_dogs['buurt_number'].astype(str).str.replace('14', '')
data_dogs = data_dogs.drop('2016', axis=1)

print(data_dogs.shape)
data_dogs.head()
```

```python
data_population = pd.read_csv("Groningen_population_data_2018.csv", sep=';') 
data_population[['buurt_number', 'buurt_name']] = data_population['Buurten'].str.extract('(?P<buurt_number>[\d]+) (?P<buurt_name>[\s\S]+)', expand=True)
data_population = data_population.drop('Buurten', axis=1)

print(data_population.shape)
print(data_population.columns)
data_population.head()
```

Let's then merge data_participatiewet, data_dogs and data_population together to the main dataset based on buurt_number column

```python
data = pd.merge(data_dogs, data_participatiewet, on="buurt_number")
data = pd.merge(data, data_population, on="buurt_number")

data.head()
```

```python
print(data.shape)
print(data.columns)
```

We can see that a lot of columns contains "|2018" in the name, but all the data in this dataset is from 2018 year, so we can remove this part od the name. And add '_' instead of space

```python
data = data.rename(columns={'aantal mensen met een uitkering Participatiewet (Pw)|2018': 'aantal mensen met een uitkering Participatiewet'})

for name, _ in data.items():
    new_name = name.replace('|2018', '')
    new_name = new_name.replace('.', '')
    new_name = new_name.replace(',', '')
    new_name = new_name.replace(':', '')
    new_name = new_name.replace('% ', 'perc ')
    new_name = new_name.replace('(', '')
    new_name = new_name.replace(')', '')
    new_name = new_name.replace(' ', '_')
    new_name = new_name.replace('_t/m_', '_')
    new_name = new_name.replace('af/uit', 'af_uit')
    data = data.rename(columns={name: new_name})


print(data.columns)
data.head()
```

```python
print(data.dtypes)
```

We can see that there are a lot of columns which should be numiric one, but they has a type 'object', so it means something wrong with their values. Let's explore it and maybe clean of fill the missing values. 

We'll start with 'aantal_mensen_met_een_uitkering_Participatiewet' column

```python
print(data['aantal_mensen_met_een_uitkering_Participatiewet'].value_counts())
print(data['gemiddeld_huishoudensinkomen'].value_counts())
```

We can see that there are 'X' and '-' values which definately not a number! So let's replace this value with 0

```python
data.replace(to_replace="X", value="0", inplace=True)
data.replace(to_replace="-", value="0", inplace=True)

for col_name in data.select_dtypes(include=[object]):
    data[col_name] = data[col_name].str.replace(',','.')
    
data.fillna(0, inplace=True)

#data.head()
```

```python
print(data.isnull().sum())
```

```python
cols=[i for i in data.columns if i not in ["buurt_name","buurt_number"]]
for col in cols:
    print("column: " + col)
    data[col]=pd.to_numeric(data[col])

print(data.dtypes)
```

```python
print(data.columns)
data.head()
```

```python
data['politie_report_aantal'] = data['aangiften_diefstal_af_uit_woning'] + data['aangiften_diefstal_auto'] + data['aangiften_diefstal_af_uit_bedrijf'] + data['aangiften_diefstal_fiets'] + data['aangiften_diefstal_af_uit_auto'] + data['aangiften_van_mishandeling'] + data['aangiften_van_vernieling']
```

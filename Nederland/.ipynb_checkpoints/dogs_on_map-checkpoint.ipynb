```python
import pandas as pd
import geopandas as gpd
import folium
import matplotlib.pyplot as plt

geo_data = gpd.GeoDataFrame(gpd.read_file("WijkBuurtkaart_2022_v1/wijkenbuurten_2022_v1.gpkg"))

geo_data.head()
```

```python
geo_data.shape
```

```python
geo_data = geo_data[geo_data['gemeentenaam'] == 'Groningen']
geo_data.set_crs(epsg='28992')
geo_data = geo_data.to_crs(epsg='4326')
geo_data.head()
```

```python
print(geo_data.shape)
```

```python
print(geo_data.columns)
```

```python
dogs_data=pd.read_csv('honden_data.csv', sep = ";")
dogs_data['buurtcode'] = dogs_data.apply(lambda row: "BU00" + str(row['buurtnr']), axis=1)
#dogs_data['2016'] = dogs_data['2016'] if dogs_data['2016']!='.' else 0
dogs_data.head()
```

```python
print(dogs_data.shape)
print(dogs_data.dtypes)
```

```python
dogs_data['2016'].value_counts()
```

```python
dogs_data['2016'] = dogs_data['2016'].replace('.',0)
convert_dict = {'2016': int}
dogs_data = dogs_data.astype(convert_dict)
print(dogs_data.dtypes)
```

```python
m = folium.Map(location=[53.22222, 6.56757], zoom_start=12, tiles='CartoDB positron')

year = '2018'
from branca.colormap import linear
colormap = linear.YlOrRd_09.scale(min(dogs_data[year]),
                                            max(dogs_data[year]))
colormap.caption = 'Amount of dogs owners in Groningen in ' + year
colormap.add_to(m)


def style_fn(feature):
    #colormap = branca.colormap.LinearColormap(["white", "green"], vmin=0, vmax=100)
    most_common = feature["properties"]["dogs_amount"]
    ss = {
        "fillColor": colormap(most_common),
        "fillOpacity": 0.9,
        "weight": 0.9,
        "opacity": 1,
        "color": "black",
    }
    return ss

def get_dogs_amount(row):
    dogs_row = dogs_data[dogs_data['buurtcode'] == row['buurtcode']]
    dogs = 0 if dogs_row.empty else dogs_row[year].iloc[0]
    return dogs

def get_age_area(row):
    max_percent = max([row['percentage_personen_0_tot_15_jaar'],
                  row['percentage_personen_15_tot_25_jaar'],
                  row['percentage_personen_25_tot_45_jaar'],
                  row['percentage_personen_45_tot_65_jaar'],
                  row['percentage_personen_65_jaar_en_ouder']])
    result = "0 tot 15 jaar" if max_percent == row['percentage_personen_0_tot_15_jaar'] else " "
    result = "15 tot 25 jaar" if max_percent == row['percentage_personen_15_tot_25_jaar'] else " "
    result = "25 tot 45 jaar" if max_percent == row['percentage_personen_45_tot_65_jaar'] else " "
    result = "45 tot 65 jaar" if max_percent == row['percentage_personen_0_tot_15_jaar'] else " "
    result = "65 jaar en ouder" if max_percent == row['percentage_personen_65_jaar_en_ouder'] else " "
    return result
                                                    

def get_marital_status(row):
    max_percent = max([row['percentage_ongehuwd'],
                  row['percentage_gehuwd'],
                  row['percentage_gescheid'],
                  row['percentage_verweduwd']])
    result = "ongehuwd" if max_percent == row['percentage_ongehuwd'] else ""
    result = "gehuwd" if max_percent == row['percentage_gehuwd'] else ""
    result = "gescheid" if max_percent == row['percentage_gescheid'] else ""
    result = "verweduwd" if max_percent == row['percentage_verweduwd'] else ""

def get_buurt_naam(row):
    return row["buurtnaam"]
    
geo_data['dogs_amount'] = geo_data.apply(lambda row: get_dogs_amount(row), axis=1)
geo_data['often_age'] = geo_data.apply(lambda row: get_age_area(row), axis=1)
geo_data['marital_status'] = geo_data.apply(lambda row: get_marital_status(row), axis=1)
geo_data['name'] = geo_data.apply(lambda row: get_buurt_naam(row), axis=1)

folium.GeoJson(
    geo_data.__geo_interface__,
    style_function=style_fn,
    popup=folium.GeoJsonPopup(fields=['dogs_amount']),
    tooltip=folium.features.GeoJsonTooltip(['name', "dogs_amount", "often_age", 'marital_status']),
).add_to(m)

m
```

```python
import pandas as pd 
data_national = pd.read_csv('migrants_in_years.csv', sep = ';')
data_national = data_national.rename(columns={'Migratieachtergrond': 'MigratieachtergrondId'})
#data_wijk_stadsdeelcode['wijkcode'] = data_wijk_stadsdeelcode['wijkcode'].str.replace('\D', '', regex=True)
#data_wijk_stadsdeelcode = data_wijk_stadsdeelcode.drop_duplicates(subset=["wijkcode"], keep='first')
print(data_national.shape)
data_national.head()
```

```python
metadata_national = pd.read_csv('migrants_in_years_metadata.csv')
#data_wijk_stadsdeelcode = data_wijk_stadsdeelcode[['Wijkcode', 'Stadsdeelcode']]
#data_wijk_stadsdeelcode = data_wijk_stadsdeelcode.rename(columns={'Wijkcode': 'wijkcode'})
#data_wijk_stadsdeelcode['wijkcode'] = data_wijk_stadsdeelcode['wijkcode'].str.replace('\D', '', regex=True)
#data_wijk_stadsdeelcode = data_wijk_stadsdeelcode.drop_duplicates(subset=["wijkcode"], keep='first')

print(metadata_national.shape)
metadata_national.head()
```

```python
def get_nationaliteit(row):
    data_row = metadata_national[metadata_national['ID'] == row['MigratieachtergrondId']]
    value = "" if data_row.empty else data_row['Title'].iloc[0]
    return value
    
def get_year(row):
    data_row = metadata_national[metadata_national['ID'] == row['Perioden']]
    value = "" if data_row.empty else data_row['Title'].iloc[0]
    return value
    

data_national = data_national.copy()

data_national['Migratieachtergrond'] = data_national.apply(lambda row: get_nationaliteit(row), axis=1)
data_national['Jaar'] = data_national.apply(lambda row: get_year(row), axis=1)

print(data_national.shape)
data_national.head()
```

```python
data_national = data_national[['MigratieachtergrondId','Migratieachtergrond', 'Jaar', 'Bevolking_1']]
data_national.to_csv(r'amsterdamers_nationaliteit.csv', index = False)
data_national.to_excel('amsterdamers_nationaliteit.xlsx', index = None, header=True)
```

```python
import pandas as pd 

new_data = pd.read_csv('buurten.csv', sep = ";")
print(new_data.shape)
new_data.head()
```

```python
data_buurt = new_data[new_data['gemeentenaam'] == "Amsterdam"]
print(data_buurt.shape)
data_buurt.head()
```

```python
data_buurt.columns.to_list()
```

```python
data = data_buurt.copy()
data.drop('fid', axis=1, inplace = True)
data.drop('geom', axis=1, inplace = True)
data.drop('gemeentecode', axis=1, inplace = True)
data.drop('gemeentenaam', axis=1, inplace = True)
data.drop('indelingswijziging_wijken_en_buurten', axis=1, inplace = True)
data.drop('water', axis=1, inplace = True)
data.drop('meest_voorkomende_postcode', axis=1, inplace = True)
data.drop('dekkingspercentage', axis=1, inplace = True)
data.drop('omgevingsadressendichtheid', axis=1, inplace = True)
data.drop('stedelijkheid_adressen_per_km2', axis=1, inplace = True)
data.drop('geboorte_totaal', axis=1, inplace = True)
data.drop('sterfte_totaal', axis=1, inplace = True)
data.drop('sterfte_relatief', axis=1, inplace = True)
data.drop('oppervlakte_totaal_in_ha', axis=1, inplace = True)
data.drop('oppervlakte_land_in_ha', axis=1, inplace = True)
data.drop('oppervlakte_water_in_ha', axis=1, inplace = True)
data.drop('jrstatcode', axis=1, inplace = True)
data.drop('jaar', axis=1, inplace = True)
```

```python
data.replace(to_replace="X", value="0", inplace=True)
data.replace(to_replace="-", value="0", inplace=True)
data.replace(to_replace=-99999999, value=0, inplace=True)

for col_name in data.select_dtypes(include=[object]):
    data[col_name] = data[col_name].str.replace(',','.')
    
data.fillna(0, inplace=True)
cols=[i for i in data.columns if i not in ["wijkcode","buurtnaam","buurtcode"]]
for col in cols:
    data[col]=pd.to_numeric(data[col])
```

```python
print(data.dtypes.to_list())
```

```python
data = data[data['buurtnaam'] != ' ']
```

```python
new_data = pd.read_csv('wijken.csv', sep = ";")
print(new_data.shape)
new_data.head()
```

```python
data_wijk = new_data[new_data['gemeentenaam'] == "Amsterdam"]
data_wijk = data_wijk[data_wijk['wijknaam'] != ' ']
data_wijk = data_wijk[['wijkcode', 'wijknaam']]
print(data_wijk.shape)
data_wijk.head()
```

```python
data = pd.merge(data, data_wijk, on="wijkcode")
print(data.shape)
```

```python
data.to_csv(r'amsterdam_buurten.csv', index = False)
```

```python
import geopandas as gpd

# reading database
geo_data = gpd.GeoDataFrame(gpd.read_file("WijkBuurtkaart_2021_v2/wijkenbuurten_2021_v2.gpkg"))
geo_data.head()
#convert Rijksdriehoeksmeting (RD) GIS coordinates format to Lat/Long 
geo_data.set_crs(epsg='28992')
geo_data = geo_data.to_crs(epsg='4326')

# get the only data we need 
geo_data = geo_data[geo_data['gemeentenaam'] == "Amsterdam"]
geo_data = geo_data[['geometry', 'buurtcode']]

# write it down to the .csv file
geo_data.to_csv(r'geodata_amsterdam.csv', index = False)
```
